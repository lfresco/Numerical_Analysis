# LAGRANGE INTERPOLATION

Dati $(n + 1)$ punti distinti $\{q_i\}_{i = 0}^n$ nell'intervallo $[0,1]$ possiamo definire l'operatore dell'interpoolazione Lagrangiana come l'operatore che dall'insieme delle funzioni continue definite sull'intervallo chiuso e limitato 0,1 ci ritorna un polinomio di ordine $n$ che soddisfa la condizione $$(L^n f)(q_i) = f(q_i)$$ per ogni $i$ che va da 0 a $n$. 

Perchè vogliamo questo operatore? Perchè ci permette di approssimare uno spazio di dimensione infinita (quello delle funzioni continue nell'intervallo ecc ecc) con uno di dimensione finita (quello dei polinomi di ordine $n$). 


Questo spazio ha dimensione $n+1$ e può essere costruito utilizzando una combinazione lineare di monomi di ordine $\le n$

```python
%matplotlib inline
from numpy import *
from pylab import *
```

Visto che dovremo andare a fare molti plot di funzioni, iniziamo con il definire uno spazio lineare che ci farà comodo. Per comodità scegliamo anche di utilizzare come numero di punti usati nel generare la funzione un numero che possa essere rappresentato in base 2

```python
ref = 1025

x = linspace(0,1,ref) # genera una sequenza di 1025 equidistanziati 
                      # nell'intervallo 0,1
n = 5  # sarà il grado del polinomio che useremo per approssimare

Pn = zeros((n+1, len(x)))  # La base di Pn 

for i in range(n + 1):
    Pn[i] = x**i  #generiamo la base del nostro spazio dei polinomi
    
figure(figsize=[10,10])
_ = plot(x,Pn.T)
    
```

Volendo utilizzare l'iterpolazione Lagrangiana per una funzione definita su $n+1$ nodi equidistanziati, dobbiamo cercare per un elemento dello spazio $P^n$ che coincida con la funzione obbiettivo in quegli specifici punti


Quindi data una base $\{p_i\}_{i=0}^n$, ogni elemento di $P^n$ può essere espresso come una combinazione lineare degli elementi della base

```python
q = linspace(0,1,n+1)  # i nodi 

A = zeros((n+1, n+1))  # la matrice che contiene i valori del poli

for j in range(n+1):
    A[:,j] = q ** j  #così facendo generiamo la base
    
f = lambda x : sin(2*pi*x)  # la funzione che vogliamo interpolare 

F = f(q)

u = linalg.solve(A,F)  # i coefficienti del polinomio 

figure(figsize=[10,5])
_ = plot(x,f(x))    # la vera funzione
_ = plot(x, Pn.T.dot(u))  # la funzione ottenuta con interpolazione
_ = plot(q, f(q), 'ro')  # i nodi
```

E' un buon metodo di interpolazione? Per vederlo dobbiamo andare a vedere il condition number di A che ci da un idea su come la nostra soluzione cambierebbe al cambiare dei punti di interpolazione. Per vedere come va proviamo ad aumenrare il numero di punti (e di conseguenza i gradi del polinomio)

```python
for i in range(3, 15):
    qtmp = linspace(0,1, i)
    Atmp = zeros((i, i))
    
    for j in range(i):
        Atmp[:, j] = qtmp ** j
        
    print("Condition number: (n=", i, " : ", linalg.cond(Atmp), ")")
```

Si può vedere come il condition number della matrice esploda all'aumentare di $n$. Quello che vogliamo quindi fare è cercare di risolvere il problema in maniera tale da tenere sotto controllo questo condition number. Un modo per farlo potrebbe essere quello di far sì che la matrice $A$ che utilizziamo sia uguale alla matrice identica. Questo equivale a scegliere come base di $P^n$ la base tale per cui $p_j(x_i) = \delta_{ij}$. Questa base viene chiamata LAGRANGE BASIS ed è costruita come $$l_i^n(x):=\prod_{j = 0, j \neq i}^n \frac{(x - x_j)}{(x_i - x_j)}$$

Per cui utilizzando questa base qua il polinomio interpolante lagrangiano possiamo ottenerlo come $$L^nf := \sum_{i = 0}^n f(x_i)l_i^n(x)$$

```python
Ln = zeros((n + 1, len(x)))

for i in range(n+1):
    Ln[i] = product([(x - q[j])/(q[i] - q[j]) for j in range(n+1) if j != i], axis = 0)
    
_ = plot(x, Ln.T)

print("Condition number: (n=", i, " : ", linalg.cond(Ln), ")")
```

Per cui l'interpolazione nei punti di nostro interessa sarà semplicemente data da

```python
y = Ln.T.dot(f(q))

figure(figsize = [10,5])
_ = plot(x, f(x))   # la vera funzione
_ = plot(x, y)      # il nostro polinomio di Lagrange
_ = plot(q, f(q), 'ro')  # i nodi
```

Proviamo adesso a vedere come si comporta con altre funzioni.

```python
def my_plot(f):
    figure(figsize = [10,5])
    y = Ln.T.dot(f(q))
    _ = plot(x, f(x))
    _ = plot(x, y)
    _ = plot(q, f(q), 'ro')
    show()
    
    
my_plot(sin)
my_plot(cos)

my_plot(lambda x : cos(2 * pi * x))

runge = lambda x : 1.0/(1 + 50*(.5-x)**2)
my_plot(runge)

```
